### Overfitting

```python
training_data, validation_data, test_data = load_data_wrapper()
training_data = list(training_data)

net2 = Network([784, 30, 10], cost=CrossEntropyCost)
net2.large_weight_initializer()
net2.SGD(training_data[:1000], 400, 10, 0.5, evaluation_data=test_data, monitor_evaluation_accuracy=True, monitor_training_cost=True)
```

### Regularization

```python
training_data, validation_data, test_data = load_data_wrapper()
training_data = list(training_data)

net3 = Network([784, 30, 10], cost=CrossEntropyCost)
net3.large_weight_initializer()
net3.SGD(training_data[:1000], 400, 10, 0.5, 
        evaluation_data=test_data, 
        lmbda=0.1, 
        monitor_evaluation_cost=True,
        monitor_evaluation_accuracy=True, 
        monitor_training_cost=True,
        monitor_training_accuracy=True
       )
```
